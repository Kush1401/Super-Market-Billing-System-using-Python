```python
# Super Market Billing System:

while True:
    amount = 0
    item_total = 0
    cold_drink = 50
    chips = 20
    ice_cream = 25
    biscuit = 10
    chocolate = 100

    # List to store billing details.
    bill_items = []

    # Taking Customer's name as input.
    name = input("Enter Customer's name: ")
    
    # Taking Customer's Contact Number as input.
    phone_number = input("Enter Customer's contact number: ")  
    
    # Taking Customer's Card type as input.
    membership_card = input("Enter type of membership card Customer is using (Golden/Silver/Bronze/None): ")

    print("\nSelect Item to add from the list:")
    print("1. Cold Drink = ₹50")
    print("2. Chips = ₹20")
    print("3. Ice Cream = ₹25")
    print("4. Biscuit = ₹10")
    print("5. Chocolate = ₹100")

    # Condition to add Different items in the Cart.
    while True:
        item = input("\nEnter the Item Code you want to add (1-5): ")

        if item == "1":
            quantity = int(input("Enter quantity of Cold Drink: "))
            item_total = cold_drink * quantity
            amount += item_total
            bill_items.append(["Cold Drink", quantity, cold_drink, item_total])

        elif item == "2":
            quantity = int(input("Enter quantity of Chips: "))
            item_total = chips * quantity
            amount += item_total
            bill_items.append(["Chips", quantity, chips, item_total])

        elif item == "3":
            quantity = int(input("Enter quantity of Ice Cream: "))
            item_total = ice_cream * quantity
            amount += item_total
            bill_items.append(["Ice Cream", quantity, ice_cream, item_total])

        elif item == "4":
            quantity = int(input("Enter quantity of Biscuit: "))
            item_total = biscuit * quantity
            amount += item_total
            bill_items.append(["Biscuit", quantity, biscuit, item_total])

        elif item == "5":
            quantity = int(input("Enter quantity of Chocolate: "))
            item_total = chocolate * quantity
            amount += item_total
            bill_items.append(["Chocolate", quantity, chocolate, item_total])

        else:
            print("Invalid item code. Please try again.")

        # Condition to ask Cashier if they want to add more items or not.
        next_item = input("Add more items? (yes/no): ")
        if next_item.lower() != "yes":
            break

    # Saving Subtotal before any discounts in a variable.
    subtotal = amount

    # Applying discount based on membership card.
    if membership_card.lower() == "golden":
        amount *= 0.80          # 20% off for Card type as "Golden".
    elif membership_card.lower() == "silver":
        amount *= 0.90          # 10% off for Card type as "Silver".
    elif membership_card.lower() == "bronze":
        amount *= 0.95          # 5% off for Card type as "Bronze".

    # Calculating total discount based on Card type.
    discount_applied = subtotal - amount

    # Code Block to display bill.
    print("\n" + "*" * 40)
    print(" " * 10 + "Supermarket Bill Receipt")
    print("*" * 40)
    print("Customer Name:        ", name)
    print("Contact Number:       ", phone_number)
    print("Membership Card Type: ", membership_card)
    print("-" * 40)

    # Print table header for items.
    print("{:<15} {:<10} {:<10} {:<10}".format("Item", "Qty", "Price", "Total"))
    print("-" * 40)

    # Print Individual item's details.
    for item in bill_items:
        print("{:<15} {:<10} {:<10} {:<10}".format(item[0], item[1], f"₹{item[2]}", f"₹{item[3]}"))

    print("-" * 40)
    print(f"Subtotal:             ₹{subtotal:.2f}")
    print(f"Discount Applied:     ₹{discount_applied:.2f}")
    print(f"Final Amount Due:     ₹{amount:.2f}")
    print("*" * 40)

    # Condition to ask after billing a Customer that if we want to bill more Customers.
    next_person = input("\nAdd next person for Billing? (yes/no): ")
    if next_person.lower() == "no":
        break
```
